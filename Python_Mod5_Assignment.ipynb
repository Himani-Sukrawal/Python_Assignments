```python
# Example of Positional Arguments -
def Personal_Info(name, age, work_place):
    print(f"{name} is {age} years old and works in {work_place}")

Personal_Info("Himani", 26, "Optum Global Solutions.")
```

```python
# Example of Keyword Arguments -
def Personal_Info(name, age, work_place):
    print(f"{name} is {age} years old and works in {work_place}")

Personal_Info(age = 26, work_place = "Optum Global Solutions.", name = "Himani")
```

```python
# Example of Default Arguments -
def Personal_Info(name, age, work_place = "Optum Global Solutions."):
    print(f"{name} is {age} years old and works in {work_place}")

Personal_Info("Himani", 26)
```

```python
# Example of Variable Length Arguments -
def Personal_Info(*args):
    print(f"{args[0]} is {args[1]} years old and works in {args[2]}")

Personal_Info("Himani", 26, "Optum Global Solutions.")
```

```python
'''
Q6 - Explain the concept of generators in Python and how they are defined.

Ans - Generator is a type of function that returns values one at a time, using the yield keyword instead of return.
Generators saves memory, are efficient for large datasets and computes the values only when needed.
It can be defined just like a normal function, but yield should be used instead of return.

'''
# Example -
def square_num(n):
  for i in range(n):
    yield i * 2

for num in square_num(4):
  print(num)
```

```python
'''
Q7 - What are the advantages of using generators over regular functions?

Ans - Advantages of using generators over regular functions are -
1. Memory efficiency as generators don’t store all values in memory. They yield one value at a time.
2. Can be executed on demand. Code runs only when needed.
3. Ideal for large datasets or for n number of sequences.
4. Useful when the user don’t need every item right away.
5. Code looks cleaner when using generators.
6. It can produce endless data streams without crashing the memory.

 '''
```

```python
'''
Q8 -What is a lambda function in Python and when is it typically used?

Ans - Lambda is a short hand function and also known as an anonymous function. It is a pre-defined python keyword.
It can take any number of arguments, but it can only have one expression. Syntax of lambda is -
lambda argument : expression.

It is typically used for simple and quick tasks and used especially when defining a full def function feels unnecessary.

'''
# Example -
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, nums))
print(squares)
```

```python
'''
Q9 - Explain the purpose and usage of the 'map()' function in Python.

Ans - Map executes a specified function for each of the item of an iterable and return a new iterator with the results.
It is more concise and efficient than writing for loops for simple transformations.
Syntax of Map is - map(function, iterable).

'''
# Example of Map using a lambda function -
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))
```

```python
# Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even(b):
  n = 0
  for i in b:
    if i % 2 == 0:
      n += i
  return n

sum_of_even([40,11,42,18])
```

```python
# Q3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_list(y):
  result = []
  for i in y:
    result.append(i ** 2)
  return result

square_list([11, 22, 33, 44, 55])
```

```python
# Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def prime_num(n):
  if n > 200:
    return "Number is above 200"
  if n <= 1:
    return False
  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False
  return True

prime_num(3)
```

```python
# Q5 - Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fib:
  def __init__(self, n_terms):
    self.n_terms = n_terms
    self.count = 0
    self.a, self.b = 0, 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.count >= self.n_terms:
      raise StopIteration

    if self.count == 0:
      self.count += 1
      return 0
    elif self.count == 1:
      self.count += 1
      return 1

# Generate the next Fibonacci number
    next_value = self.a + self.b
    self.a, self.b = self.b, next_value
    self.count += 1
    return next_value

abc = Fib(6)

for num in abc:
    print(num)
```

```python
# Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(n):
  for i in range(n + 1):
    yield 2 ** i

print(list(power_of_two(8)))
```

```python
# Q7. Implement a generator function that reads a file line by line and yields each line as a string.

with open("try.txt", "w") as f:
  f.write("This is line 1\n")
  f.write("This is line 2\n")
  f.write("This is line 3\n")

def file(filename):
  with open(filename, 'r') as f:
    for i in f:
      yield i.strip()

print(list(file("try.txt")))
```

```python
# Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

list_of_tuple = [(1, 2), (2, 1), (4, 3), (6, 2)]
sorted_list = sorted(list_of_tuple, key = lambda x : x[1])
print(sorted_list)
```

```python
# Q9. Write a Python program that uses 'map()' to convert a list of temperatures from Celsius to Fahrenheit.

Celsius = [10, 12, 24, 44]
Fahrenheit = list(map(lambda x : (x * 9/5) + 32, Celsius))
print("Celsius Values :", Celsius)
print("Fahrenheit Values :", Fahrenheit)
```

```python
# Q10. Create a Python program that uses 'filter()' to remove all the vowels from a given string.

vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
string1 = input("Enter a string here :")
filtered_string = ''.join(filter(lambda x : x not in vowels, string1))

print("Here is the string without the vowels:", filtered_string)
```
